```python
import matplotlib.pyplot as plt
import numpy

%matplotlib inline
```

```python
def maprange(x, in_min, in_max, out_min, out_max):
  return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min
```

```python
distances = list(range(20, 110, 10))
outputs = [490, 375, 300, 230, 200, 170, 155, 135, 125]
voltages = [maprange(x, 0, 1023, 0, 5) for x in outputs]

ref_distances = [0, 10, 15] + list(range(20, 160, 10))
ref_voltages = [0.01, 2.3, 2.75, 2.5, 2, 1.53, 1.25, 1.05, 0.9, 0.8, 0.72, 0.66, 0.6, 0.55, 0.5, 0.49, 0.45]
ref_outputs = [maprange(x, 0, 5, 0, 1023) for x in ref_voltages]

[A, B] = numpy.polyfit(numpy.log(distances), outputs, 1)
print(A, B)

print([maprange(x, 490, 230, 20, 50) for x in outputs])

plt.plot(distances, outputs, 'b-')
plt.plot(ref_distances, ref_outputs, 'r-')
plt.ylim([0, 650])
plt.xlim([0, 150])
plt.show()
```
